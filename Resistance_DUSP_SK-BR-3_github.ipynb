```python
import boolean2 as b2
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
%matplotlib inline
```

Model Activation of DUSP4 (inhibits ERK, JNK; activated by ERK)

```python
model_definition = """
# Initial values
Herceptin = True
HER2 = True
ERK12 = True
JNK12 = False
P38 = False
Survival = True
DUSP4 = True
A = True

# Update rules
HER2* = not Herceptin
ERK12* = HER2 or not DUSP4
JNK12* = HER2 or not DUSP4
P38* = HER2 
Survival* = ERK12 or not(JNK12 and P38)
DUSP4* = ERK12 or A
"""

model = b2.Model(text=model_definition, mode='sync')
model.initialize()
model.iterate(steps=6)

states = dict()
image = list()

print model.data
for node in sorted(model.data):
    states[node] = model.data[node]
    image.append(states[node])

plt.figure(figsize=(8, 8))
plt.yticks(range(0, 9), sorted(model.data), fontsize=20)
plt.xticks(fontsize=20)
plt.imshow(image, cmap=plt.cm.get_cmap('RdYlGn'), interpolation='none')

green = mpatches.Patch(color='green', label='ON')
red = mpatches.Patch(color='firebrick', label='OFF')
plt.legend(handles=[green, red])
plt.savefig('DUSP4_model1ACT', dpi=90)
```
