**DUSP16 model with ERK activation**

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
def DUSP16model_d1(x, t):

    # Definition of proteins
    Herceptin = 100
    HER2ac = x[0]
    HER2i = x[1]
    ERK12ac = x[2]
    ERK12i = x[3]
    JNK12ac = x[4]
    JNK12i = x[5]
    P38ac = x[6]
    P38i = x[7]
    DUSP16ac = x[8]
    DUSP16i = x[9]
    Survival = x[10]
    
    # Initialisation of parameters
    k1 = 1
    k2 = 1
    k3 = 1
    k4 = 1
    k5 = 1
    d1 = 1
    d2 = 1
    d3 = 1
    d4 = 1
    d5 = 1
    s1 = 1
    s2 = 1
    
    # Differential equations
    HER2acd = k1*HER2i - d1*HER2ac*Herceptin
    HER2id = d1*HER2ac*Herceptin - k1*HER2i
    
    ERK12acd = k2*ERK12i*HER2ac - d2*ERK12ac
    
    ERK12id = d2*ERK12ac - k2*ERK12i*HER2ac
    
    JNK12acd = k3*JNK12i*HER2ac - d3*JNK12ac*DUSP16ac
    JNK12id = d3*JNK12ac*DUSP16ac - k3*JNK12i*HER2ac
    P38acd = k4*P38i*HER2ac - d4*P38ac*DUSP16ac
    P38id = d4*P38ac*DUSP16ac - k4*P38i*HER2ac
    DUSP16acd = k5*DUSP16i*ERK12ac - d5*DUSP16ac
    DUSP16id = d5*DUSP16ac - k5*DUSP16i*ERK12ac
    Survivald = s1*ERK12ac - s2*(JNK12ac*P38ac)**0.6
    
    return [HER2acd, HER2id, ERK12acd, ERK12id, JNK12acd, JNK12id, P38acd, P38id, DUSP16acd, DUSP16id, Survivald]


def DUSP16model_d10(x, t):

    # Definition of proteins
    Herceptin = 100
    HER2ac = x[0]
    HER2i = x[1]
    ERK12ac = x[2]
    ERK12i = x[3]
    JNK12ac = x[4]
    JNK12i = x[5]
    P38ac = x[6]
    P38i = x[7]
    DUSP16ac = x[8]
    DUSP16i = x[9]
    Survival = x[10]
    
    # Initialisation of parameters
    k1 = 1
    k2 = 1
    k3 = 1
    k4 = 1
    k5 = 1
    d1 = 1
    d2 = 1
    d3 = 1
    d4 = 1
    d5 = 10
    s1 = 1
    s2 = 1
    
    # Differential equations
    HER2acd = k1*HER2i - d1*HER2ac*Herceptin
    HER2id = d1*HER2ac*Herceptin - k1*HER2i
    ERK12acd = k2*ERK12i*HER2ac - d2*ERK12ac
    ERK12id = d2*ERK12ac - k2*ERK12i*HER2ac
    JNK12acd = k3*JNK12i*HER2ac - d3*JNK12ac*DUSP16ac
    JNK12id = d3*JNK12ac*DUSP16ac - k3*JNK12i*HER2ac
    P38acd = k4*P38i*HER2ac - d4*P38ac*DUSP16ac
    P38id = d4*P38ac*DUSP16ac - k4*P38i*HER2ac
    DUSP16acd = k5*DUSP16i*ERK12ac - d5*DUSP16ac
    DUSP16id = d5*DUSP16ac - k5*DUSP16i*ERK12ac
    Survivald = s1*ERK12ac - s2*(JNK12ac*P38ac)**0.6
    
    return [HER2acd, HER2id, ERK12acd, ERK12id, JNK12acd, JNK12id, P38acd, P38id, DUSP16acd, DUSP16id, Survivald]


def DUSP16model_d20(x, t):

    # Definition of proteins
    Herceptin = 100
    HER2ac = x[0]
    HER2i = x[1]
    ERK12ac = x[2]
    ERK12i = x[3]
    JNK12ac = x[4]
    JNK12i = x[5]
    P38ac = x[6]
    P38i = x[7]
    DUSP16ac = x[8]
    DUSP16i = x[9]
    Survival = x[10]
    
    # Initialisation of parameters
    k1 = 1
    k2 = 1
    k3 = 1
    k4 = 1
    k5 = 1
    d1 = 1
    d2 = 1
    d3 = 1
    d4 = 1
    d5 = 20
    s1 = 1
    s2 = 1
    
    # Differential equations
    HER2acd = k1*HER2i - d1*HER2ac*Herceptin
    HER2id = d1*HER2ac*Herceptin - k1*HER2i
    ERK12acd = k2*ERK12i*HER2ac - d2*ERK12ac
    ERK12id = d2*ERK12ac - k2*ERK12i*HER2ac
    JNK12acd = k3*JNK12i*HER2ac - d3*JNK12ac*DUSP16ac
    JNK12id = d3*JNK12ac*DUSP16ac - k3*JNK12i*HER2ac
    P38acd = k4*P38i*HER2ac - d4*P38ac*DUSP16ac
    P38id = d4*P38ac*DUSP16ac - k4*P38i*HER2ac
    DUSP16acd = k5*DUSP16i*ERK12ac - d5*DUSP16ac
    DUSP16id = d5*DUSP16ac - k5*DUSP16i*ERK12ac
    Survivald = s1*ERK12ac - s2*(JNK12ac*P38ac)**0.6
    
    return [HER2acd, HER2id, ERK12acd, ERK12id, JNK12acd, JNK12id, P38acd, P38id, DUSP16acd, DUSP16id, Survivald]


def DUSP16model_d50(x, t):

    # Definition of proteins
    Herceptin = 100
    HER2ac = x[0]
    HER2i = x[1]
    ERK12ac = x[2]
    ERK12i = x[3]
    JNK12ac = x[4]
    JNK12i = x[5]
    P38ac = x[6]
    P38i = x[7]
    DUSP16ac = x[8]
    DUSP16i = x[9]
    Survival = x[10]
    
    # Initialisation of parameters
    k1 = 1
    k2 = 1
    k3 = 1
    k4 = 1
    k5 = 1
    d1 = 1
    d2 = 1
    d3 = 1
    d4 = 1
    d5 = 50
    s1 = 1
    s2 = 1
    
    # Differential equations
    HER2acd = k1*HER2i - d1*HER2ac*Herceptin
    HER2id = d1*HER2ac*Herceptin - k1*HER2i
    ERK12acd = k2*ERK12i*HER2ac - d2*ERK12ac
    ERK12id = d2*ERK12ac - k2*ERK12i*HER2ac
    JNK12acd = k3*JNK12i*HER2ac - d3*JNK12ac*DUSP16ac
    JNK12id = d3*JNK12ac*DUSP16ac - k3*JNK12i*HER2ac
    P38acd = k4*P38i*HER2ac - d4*P38ac*DUSP16ac
    P38id = d4*P38ac*DUSP16ac - k4*P38i*HER2ac
    DUSP16acd = k5*DUSP16i*ERK12ac - d5*DUSP16ac
    DUSP16id = d5*DUSP16ac - k5*DUSP16i*ERK12ac
    Survivald = s1*ERK12ac - s2*(JNK12ac*P38ac)**0.6
    
    return [HER2acd, HER2id, ERK12acd, ERK12id, JNK12acd, JNK12id, P38acd, P38id, DUSP16acd, DUSP16id, Survivald]


def DUSP16model_d100(x, t):

    # Definition of proteins
    Herceptin = 100
    HER2ac = x[0]
    HER2i = x[1]
    ERK12ac = x[2]
    ERK12i = x[3]
    JNK12ac = x[4]
    JNK12i = x[5]
    P38ac = x[6]
    P38i = x[7]
    DUSP16ac = x[8]
    DUSP16i = x[9]
    Survival = x[10]
    
    # Initialisation of parameters
    k1 = 1
    k2 = 1
    k3 = 1
    k4 = 1
    k5 = 1
    d1 = 1
    d2 = 1
    d3 = 1
    d4 = 1
    d5 = 100
    s1 = 1
    s2 = 1
    
    # Differential equations
    HER2acd = k1*HER2i - d1*HER2ac*Herceptin
    HER2id = d1*HER2ac*Herceptin - k1*HER2i
    ERK12acd = k2*ERK12i*HER2ac - d2*ERK12ac
    ERK12id = d2*ERK12ac - k2*ERK12i*HER2ac
    JNK12acd = k3*JNK12i*HER2ac - d3*JNK12ac*DUSP16ac
    JNK12id = d3*JNK12ac*DUSP16ac - k3*JNK12i*HER2ac
    P38acd = k4*P38i*HER2ac - d4*P38ac*DUSP16ac
    P38id = d4*P38ac*DUSP16ac - k4*P38i*HER2ac
    DUSP16acd = k5*DUSP16i*ERK12ac - d5*DUSP16ac
    DUSP16id = d5*DUSP16ac - k5*DUSP16i*ERK12ac
    Survivald = s1*ERK12ac - s2*(JNK12ac*P38ac)**0.6
    
    return [HER2acd, HER2id, ERK12acd, ERK12id, JNK12acd, JNK12id, P38acd, P38id, DUSP16acd, DUSP16id, Survivald]


def DUSP16model_d500(x, t):

    # Definition of proteins
    Herceptin = 100
    HER2ac = x[0]
    HER2i = x[1]
    ERK12ac = x[2]
    ERK12i = x[3]
    JNK12ac = x[4]
    JNK12i = x[5]
    P38ac = x[6]
    P38i = x[7]
    DUSP16ac = x[8]
    DUSP16i = x[9]
    Survival = x[10]
    
    # Initialisation of parameters
    k1 = 1
    k2 = 1
    k3 = 1
    k4 = 1
    k5 = 1
    d1 = 1
    d2 = 1
    d3 = 1
    d4 = 1
    d5 = 500
    s1 = 1
    s2 = 1
    
    # Differential equations
    HER2acd = k1*HER2i - d1*HER2ac*Herceptin
    HER2id = d1*HER2ac*Herceptin - k1*HER2i
    ERK12acd = k2*ERK12i*HER2ac - d2*ERK12ac
    ERK12id = d2*ERK12ac - k2*ERK12i*HER2ac
    JNK12acd = k3*JNK12i*HER2ac - d3*JNK12ac*DUSP16ac
    JNK12id = d3*JNK12ac*DUSP16ac - k3*JNK12i*HER2ac
    P38acd = k4*P38i*HER2ac - d4*P38ac*DUSP16ac
    P38id = d4*P38ac*DUSP16ac - k4*P38i*HER2ac
    DUSP16acd = k5*DUSP16i*ERK12ac - d5*DUSP16ac
    DUSP16id = d5*DUSP16ac - k5*DUSP16i*ERK12ac
    Survivald = s1*ERK12ac - s2*(JNK12ac*P38ac)**0.6
    
    return [HER2acd, HER2id, ERK12acd, ERK12id, JNK12acd, JNK12id, P38acd, P38id, DUSP16acd, DUSP16id, Survivald]
```
